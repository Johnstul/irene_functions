```python
import numpy as np

%matplotlib inline

from matplotlib.pyplot import plot, show, cm

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
lo = -100
hi = 100

n = hi - lo + 1
# print(n)

x = np.linspace(lo, hi, n)
y = np.linspace(lo, hi, n)

def f_1(x,y):
    return np.sqrt(x**2 - y**2)

def f_2(x,y):
    return np.sqrt(y**2 - x**2)

X, Y = np.meshgrid(x, y)

Z1 = f_1(X, Y)
Z2 = f_2(X, Y)

# Ignore the box that appears in red
```

```python
# cmap = cm.jet
# cmap = cm.autumn
cmap = cm.coolwarm

fig_w = 15
fig_h = 15

smooth = 150

fig = plt.figure(figsize=(fig_w,fig_h))
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z1, smooth, cmap = cmap)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
fig.suptitle('3D Plot of z = sqrt(x^2 - y^2)', fontsize = 20)

# elevation refers to n degrees above the x-y plane
elevation = 60
# azimuth refers to rotation n degrees counter-clockwise about the z-axis
azimuth = 30
ax.view_init(elevation, azimuth)

plt.show()

#################

fig = plt.figure(figsize=(fig_w,fig_h))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z2, smooth, cmap = cmap)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
fig.suptitle('3D Plot of z = sqrt(y^2 - x^2)', fontsize = 20)

# elevation refers to n degrees above the x-y plane
elevation = 60
# azimuth refers to rotation n degrees counter-clockwise about the z-axis
azimuth = 30
ax.view_init(elevation, azimuth)

plt.show()
```
